```python
# Looking at Average Pedestal Value (First 10%) on Event by Event Basis

recordLength = int(data[0])
holderArray = []
holdEnd = []

frameCounter = 0

for nthframe in range(len(data)/(recordLength+7)): # for each frame
    ped_ini_10pct = 0. # initialize 10% ped variable
    ped_end_10pct = 0.
#     print(ped_10pct, nthframe)
    for val in data[nthframe*(recordLength+7)+7:nthframe*(recordLength+7)+7 + recordLength/10]:
        ped_ini_10pct += float(val)
#     print(ped_ini_10pct, nthframe)
    for val in data[nthframe*(recordLength+7)+7 + 9*recordLength/10:(nthframe+1)*(recordLength+7)-1]:
        ped_end_10pct += float(val)
#     print(ped_end_10pct, nthframe)
    holderArray.append(ped_ini_10pct/(recordLength/10))
    holdEnd.append(ped_end_10pct/(recordLength/10))
    
print(holderArray)
print(holdEnd)

plt.plot(np.array(holderArray), np.array(holdEnd), marker='o', lw='0')
# plt.plot(holdEnd)
plt.show()
```

```python
recordLength = int(data[0])
holderArray = []
charge = []
voltage = 0.0

frameCounter = 0

n = 9 # nth frame
ini = n*1037+7 # nth frame + header for (n+1)th frame
j = ini # initialize j with ini

sum_event = 0.
for val in data[ini:ini+100]:
    sum_event += float(val)
# print(sum_event)

while j < ini+1030:
    holderArray.append((float(data[j]) - sum_event/len(data[ini:ini+100]))/8192.)
    j+=1
    
    if j%(recordLength + 7) == 0:
        voltage = sum(holderArray)
        j += 7
    
        plt.plot(holderArray)
        plt.title("v = " + str(voltage) + ", q = " + str(-(voltage)*2E-9/50.))
        plt.xlabel("time (s)")
        plt.ylabel("voltage (V)")
        plt.show()

print("voltage = " + str(voltage))
print("charge = " + str(-(voltage)*2E-9/50.) + "\n")
print(holderArray)
```
