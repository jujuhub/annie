```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
%matplotlib inline
```

```python
with open("my_charges.txt") as f1:
    charge = np.loadtxt(f1, dtype = float)
```

```python
print(charge)
print(len(charge))
print(type(charge))
```

```python
"""
for q in charge:
    if q < -0.5e-11:
        print(q)
        q = -1000
"""

i=0
new_charge = []

while (i < len(charge)):
    if charge[i] < -0.5e-11:
        print(charge[i])
    else:
        new_charge.append(charge[i])
    i+=1

print(len(new_charge))
print(type(new_charge))

new_charge = np.array(new_charge)

print(type(new_charge))
print(new_charge)
```

```python
"""
    plotting
"""
fig = plt.figure()
ax1 = fig.add_subplot(111)

y_data, bins, patches = ax1.hist(new_charge, bins=100, facecolor='orange')


plt.yscale('log', nonposy='clip')
#plt.xlabel('Coulombs')
#plt.ylabel('Counts')
```

```python
bins = bins*1e12
#print(bins)
bin_centers = np.array(bins[:-1]+(bins[1]-bins[0])/2.0)
```

```python
# bin uncertainties
uncert_data = []
for i in range(y_data.size):
    uncert_data.append(np.sqrt(y_data[i]))
    if uncert_data[i] == 0:
        uncert_data[i] = 1
```

```python
"""
    fitting a double gaussian
"""
def double_gauss(x, mu1, sigma1, n1, mu2, sigma2, n2):
    dbgauss = (n1/(np.sqrt(2*np.pi)*sigma1))*np.exp(-(x-mu1)**2/(2*sigma1**2)) 
            + (n2/(np.sqrt(2*np.pi)*sigma2))*np.exp(-(x-mu2)**2/(2*sigma2**2))
    return dbgauss
```

```python
dbgauss_guess = [0, 1, 1000, 1, 2, 100]
popt, pcov = curve_fit(double_gauss, bin_centers, y_data, p0=dbgauss_guess, sigma=uncert_data)
#popt2, pcov2 = curve_fit(double_gauss, bin_centers, y_data, p0=dbgauss_guess)
```

```python
print(popt)
#print(popt2)
```

```python
dbgauss_fit = double_gauss(bin_centers, *popt)
bin_centers = bin_centers*1e-12
```

```python
plt.plot(bin_centers, dbgauss_fit)
plt.yscale('log')
plt.show()
```

### Gain curve

```python
# voltages = np.array([1304, 1405, 1508])
# gain = np.array([2.06E7, 4.98E7, 1.66E8])

voltages = np.array([1420, 1550, 1450, 1475])
gain = np.array([0.66E7, 2.25E7, 0.74E7, 0.86E7])
```

```python
def gain_curve(volt, m, b):
    return m*volt + b

# def gain_curve(volt, a, m, b):
#     return a*np.exp(m*volt) + b

popt_gain, pcov_gain = curve_fit(gain_curve, voltages, np.log10(gain))
print(popt_gain)

x = np.arange(1000, 1600, 1)
gain_fit = gain_curve(x, *popt_gain) # log of the gain
# print(gain_fit)

plt.figure(figsize=(10,6))
plt.title('Gain Curve', fontsize=15)
plt.plot(x, gain_fit, label='fit')
plt.plot(voltages, np.log10(gain), marker='o', lw=0, label='data')
plt.plot(np.array([1000, 1600]), np.array([np.log10(1e7), np.log10(1e7)]), 'k-', label='g = 1E7')
plt.xlabel('voltage (V)')
plt.ylabel('log(gain)')
# plt.yscale('log')
plt.legend(loc='best')
# plt.savefig("gainCurve.png")
```
