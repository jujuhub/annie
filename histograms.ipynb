### Peak-to-valley

```python
with open('./P2V/peak2valley.txt') as f:
    peak2valley = np.loadtxt(f, dtype=float)

with open('./P2V/ET_peak2valley.txt') as f1:
    et_p2v = np.loadtxt(f1, dtype=float)
    
with open('./P2V/WB_peak2valley.txt') as f2:
    wb_p2v = np.loadtxt(f2, dtype=float)

plt.figure(figsize=(9,6))
#y_peak, peak_bin_edges, peak_patches = plt.hist(peak2valley, bins=15, range=(0,3), 
#                                                alpha=0.75, facecolor='orange')
et_pk, etp_bin_edges, etp_patches = plt.hist(et_p2v, bins=20, range=(0,4), 
                                                alpha=0.5, facecolor='blue', label='ET/EL')
wb_pk, wbp_bin_edges, wbp_patches = plt.hist(wb_p2v, bins=20, range=(0,4), 
                                                alpha=0.5, facecolor='green', label='WB')

plt.title("Peak-to-Valley Ratio Distribution", fontsize=15)
plt.xlabel("peak-to-valley", fontsize=13)
plt.ylabel("# of PMTs", fontsize=13)
plt.ylim(0,20)
plt.legend(loc=0)
#plt.savefig('peak2valley.png')
```
